# Histogram, PDF, CDF

```python
# What about 1-D scatter plot using just one feature?
#1-D scatter plot of petal-length
import numpy as np
iris_setosa = iris.loc[iris["species"] == "setosa"]
iris_virginica = iris.loc[iris["species"] == "virginica"]
iris_versicolor = iris.loc[iris["species"] == "versicolor"]

plt.plot(iris_setosa["petal_length"], np.zeros_like(iris_setosa['petal_length']), 'o')
plt.plot(iris_versicolor["petal_length"], np.zeros_like(iris_versicolor['petal_length']), 'o')
plt.plot(iris_virginica["petal_length"], np.zeros_like(iris_virginica["petal_length"]), 'o')

plt.show()
```

```python
# Univariate Analysis ON PDF -> One Variable analysis

# use case: How to find out which of four SL, PL, PW, SW is more 
# useful in classification of flowers species

"""
    Out of all four histograms above, take the petal_length
    as it really separates from others
"""
```

```python
# Need for Cumulative Distribution Function (CDF)
# We can do flower classification more clearly when we can get CDF which provides more weightage to the data
```

```python
counts_s, bin_edges_s = np.histogram(iris_setosa['petal_length'], bins=10, density=True)
pdf_s = counts_s/(sum(counts_s))
count_vi, bin_edges_vi = np.histogram(iris_virginica['petal_length'], bins=10, density=True)
pdf_vi = count_vi/(sum(count_vi))
count_ve, bin_edges_ve = np.histogram(iris_versicolor['petal_length'], bins=10, density=True)
pdf_ve = count_ve/(sum(count_ve))
```

```python
cdf_s = np.cumsum(pdf_s)
cdf_ve = np.cumsum(pdf_ve)
cdf_vi = np.cumsum(pdf_vi)

plt.plot(bin_edges_s[1:], pdf_s)
plt.plot(bin_edges_s[1:], cdf_s)

plt.plot(bin_edges_ve[1:], pdf_ve)
plt.plot(bin_edges_ve[1:], cdf_ve)

plt.plot(bin_edges_vi[1:], pdf_vi)
plt.plot(bin_edges_vi[1:], cdf_vi)

plt.show()
```

```python
# By looking at above grid, we can conclude following
# we can say that, 100% flowers are setosa with petal_length<=2
# there is 95% chance that flower is virginica with petal_length > 3.2 && petal_length < 5 ( 5% chance is that it could be versicolor)
# ther is 10% chance that flower is versicolor with pertal length <=5
```

# Mean, Variance and standard-deviation

```python
print(np.mean(iris_setosa['petal_length']))
print(np.mean(iris_virginica['petal_length']))
print(np.mean(iris_versicolor['petal_length']))

print(np.mean(np.append(iris_setosa['petal_length'], 50)))
```

```python
# variance is average distance from mean which is used to calculate
# spread in histograms
# variance = 1/n*sum(xi-mean)^2
# standard deviation = sqrt(variance) -> which is called spread in plot
```

```python
print(np.std(iris_setosa['petal_length']))
print(np.std(iris_virginica['petal_length']))
print(np.std(iris_versicolor['petal_length']))
```
