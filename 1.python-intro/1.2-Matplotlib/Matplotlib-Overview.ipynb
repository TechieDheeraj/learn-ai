```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,3,4,1]) # treat numbers as y-axis and indices as x-axis
plt.ylabel("Numbers") # label y-axis
plt.xlabel("Values")  # label x-axis
plt.title("My Plot")  # title of plot
plt.grid()  # to show the grid
plt.show()
```

```python
plt.plot([1,2,3,4], [1,4,9,16]) # first list is x-axis and second list is y-axis
plt.ylabel("Squares")
plt.xlabel("Numbers")
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4], [1, 4, 9, 16], 'ro')
plt.ylabel("squares")
plt.xlabel("numbers")
plt.grid()
plt.show()
```

```python
import numpy as np

t = np.arange(0., 5., 0.2)

plt.plot(t, t**2, 'b--', label='^2')
plt.plot(t, t**2.2, 'rs', label='^2.2')
plt.plot(t, t**2.5, 'g^', label='^2.5')
plt.legend()

plt.show()
```

```python
x = [1,2,3,4]
y = [3,4,6,7]
plt.plot(x, y, linewidth=5.0) # increase line width
plt.grid()
plt.show()
```

```python
x1=[1, 2, 3, 4]
y1=[1, 4, 9, 16]
x2=[1, 2, 3, 4]
y2=[2, 4, 6, 8]

lines = plt.plot(x1, y1, x2, y2)

plt.setp(lines[0], color='r', linewidth=2.0)
plt.setp(lines[1], 'color', 'g', 'linewidth', 2.0) # can also use matlab syntax
plt.grid()
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)   # In Figure: 2(rows), 1(one column), 1st plot
plt.grid()
plt.plot(t1, f(t1), 'b-')

plt.subplot(212)   # In Figure: 2(rows), 1(one column), 2nd plot
plt.grid()
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1,2,3])
plt.subplot(212)
plt.plot([2,4,6])

plt.figure(2)
plt.plot([7,8,9])

plt.figure(1)
plt.subplot(211)
plt.title('Easy as 1, 2, 3')
plt.grid()
plt.show()
```

```python
t1 = np.arange(0., 5., 0.2)
t2 = np.arange(0., 5., 2.0)

plt.figure(1)
plt.subplot(221)
plt.plot(t1, t1**2, 'r--')
plt.subplot(222)
plt.plot(t1, np.pi*t1, 'g^')
plt.figure(2)
plt.subplot(223)
plt.plot(t2, t2**2, 'b--')
plt.subplot(224)
plt.plot(t2, np.pi*t2, 'rs')
plt.show()
```
